# 📊 Chicago Crime Dataset – Column Descriptions

## 🆔 Identifiers

### Unnamed: 0
- Extra index column created during data export
- ❌ Not useful → should be dropped

### ID
- Unique identifier for each crime record
- Used to distinguish entries
- ❌ Not useful for modeling (no predictive value)

### Case Number
- Unique case reference used by police
- May have slight missing values
- ❌ Not useful for modeling

---

## 🕒 Date & Time

### Date
- Date and time when the crime occurred
- Useful for extracting:
  - hour
  - day
  - month

### Updated On
- Last time the record was updated
- Not very useful for prediction
- Can be dropped or ignored

### Year
- Year of the crime occurrence
- Can be treated as a time feature

---

## 📍 Location (Text-Based)

### Block
- Approximate street address (e.g., "001XX W MADISON ST")
- Represents area, not exact location

### Location Description
- Type of location where crime occurred
- Examples: STREET, RESIDENCE, SCHOOL

### Location
- Combined Latitude & Longitude as text
- ❌ Redundant (already have Lat/Long)

---

## 📍 Location (Geographic)

### Latitude
- North-South geographic coordinate
- Continuous numerical feature

### Longitude
- East-West geographic coordinate
- Continuous numerical feature

### X Coordinate
- Local map projection coordinate (horizontal)
- ❌ Redundant with Latitude/Longitude

### Y Coordinate
- Local map projection coordinate (vertical)
- ❌ Redundant with Latitude/Longitude

---

## 🏙️ Administrative / Region-Based

### Beat
- Smallest police patrol area
- Used by police for operations
- Categorical

### District
- Police district where crime occurred
- Larger than Beat
- Categorical

### Ward
- Political division of Chicago
- Used for elections and governance
- Categorical

### Community Area
- Neighborhood grouping (77 areas)
- Used for analysis and statistics
- Categorical

---

## 🚔 Crime Details

### IUCR
- Illinois Uniform Crime Reporting code
- Encodes type of crime

### Primary Type
- Main category of crime
- Example: THEFT, BATTERY, ASSAULT

### Description
- Detailed description of the crime
- More specific than Primary Type

### FBI Code
- Crime classification used by the FBI
- Groups crimes into broader categories

---

## ⚖️ Crime Characteristics

### Arrest
- Whether an arrest was made
- Binary (True/False)

### Domestic
- Indicates if crime was domestic-related
- Binary (True/False)

---

## 🧠 Summary Notes

- Many columns describe **location in different ways**
- Some features are **redundant**:
  - Latitude/Longitude vs X/Y vs Location
- Some numeric columns are actually **categorical**:
  - Beat, District, Ward, Community Area
- Some columns should be **dropped**:
  - Unnamed: 0, ID, Case Number, X/Y, Location

---

## 2 -  Handling missing values

````python
df.isnull().sum()
````

### 1 - Categorical Values

````python
# calculate the percentage of missing values for each categorical column
cat_cols = [
    'Date',
    'Block',
    'Primary Type',
    'Description',
    'Location Description',
    'Arrest',
    'Domestic',
    'Beat',
    'District',
    'Ward',
    'Community Area',
    'Year',
    'Updated On',
    'Location'
]

missing_percent_cat = df[cat_cols].isnull().mean() * 100
print(missing_percent_cat)
````

````python
#columns with more than 0% missing values in categorical columns
cols_with_missing_cat = missing_percent_cat[missing_percent_cat > 0].index.tolist()
cols_with_missing_cat
````

````python
# handling them by taking the mode of each column and filling the missing values with it
for col in cols_with_missing_cat:
    mode_value = df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)
````

````python
#checking if there are still missing values in categorical columns
missing_percent_cat_after = df[cat_cols].isnull().mean() * 100
missing_percent_cat_after
````

### 2 - Numerical values

````python
df[['Latitude', 'Longitude']].isnull().sum() # only real numerical columns
````

````python
df['Community Area'].isnull().sum()
````

### We will now use the " Community area " column to fill the missing values in "latitude" & "longitude"

````python
df['Latitude'] = df.groupby('Community Area')['Latitude'].transform(
    lambda x: x.fillna(x.median())
)
````

````python
df['Longitude'] = df.groupby('Community Area')['Longitude'].transform(lambda x: x.fillna(x.median()))
````

````python
def group_location_final(loc):
    if pd.isnull(loc):
        return 'Other'

    loc = loc.upper()

    # 🏠 Residential
    if any(word in loc for word in [
        'RESIDENCE', 'APARTMENT', 'HOUSE', 'HOME', 'PORCH',
        'YARD', 'GARAGE', 'BASEMENT', 'ROOMING', 'CHA APARTMENT',
        'RESIDENTIAL'
    ]):
        return 'Residential'

    # 🏬 Commercial
    elif any(word in loc for word in [
        'STORE', 'SHOP', 'RETAIL', 'RESTAURANT', 'BAR', 'TAVERN',
        'BANK', 'OFFICE', 'BUSINESS', 'HOTEL', 'MOTEL',
        'GAS STATION', 'DRUG STORE', 'CURRENCY EXCHANGE',
        'DEALERSHIP', 'LAUNDROMAT', 'CLEANERS', 'PAWN'
    ]):
        return 'Commercial'

    # 🚗 Transportation
    elif any(word in loc for word in [
        'CTA', 'TRAIN', 'BUS', 'PLATFORM', 'TAXI',
        'VEHICLE', 'AIRPORT', 'AIRCRAFT', 'BOAT',
        'TRUCK', 'HIGHWAY', 'EXPRESSWAY', 'RAILROAD',
        'SUBWAY'
    ]):
        return 'Transportation'

    # 🏫 Institution
    elif any(word in loc for word in [
        'SCHOOL', 'COLLEGE', 'UNIVERSITY',
        'HOSPITAL', 'MEDICAL', 'POLICE',
        'JAIL', 'GOVERNMENT', 'CHURCH',
        'FACILITY', 'LIBRARY', 'FIRE STATION',
        'DAY CARE'
    ]):
        return 'Institution'

    # 🌳 Public / Outdoor
    elif any(word in loc for word in [
        'STREET', 'SIDEWALK', 'ALLEY',
        'PARK', 'LOT', 'FOREST', 'LAKE',
        'RIVER', 'FIELD', 'PRAIRIE',
        'BRIDGE', 'DUMPSTER', 'GANGWAY'
    ]):
        return 'Public'

    # ❓ Everything else
    else:
        return 'Other'
````

````python
df['Location_Group'] = df['Location Description'].apply(group_location_final)
````

````python
df['Location_Group'].value_counts()
````

````python
df = pd.get_dummies(
    df,
    columns=['Location_Group', 'Primary Type'],
    prefix=['Loc', 'Crime']
)
````
